```python
from collections import Counter
wordDict = Counter()
f = open("518연설문.txt", "r")

sentences = f.readlines()

for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩 띄어쓰기로 분리
        wordDict[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
wordlist=[]
freqlist=[]
for word, freq in wordDict.most_common(20):
    wordlist.append(word)
    freqlist.append(int(freq))
    print(word,freq)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun gothic')
plt.barh(wordlist,freqlist)
plt.show()
```

```python
from konlpy.tag import Hannanum
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud


f = open("518연설문.txt", "r")
wordDict = Counter()
sentences = f.readlines()

for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        wordDict[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
tags=wordDict.most_common(10)

wordcloud=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='white', width=1200, height=800).generate_from_frequencies(dict(tags))

fig=plt.figure()
plt.axis('off')
plt.imshow(wordcloud)        
        

```

```python
from PIL import Image
import numpy as np

hong=np.array(Image.open('문대통령사진.jpg'))

text = open("518연설문.txt", "r").read() #encoding='utf8'는 오류가 나는 경우 사용

engin=Hannanum()
nouns=engin.nouns(text)
nouns=[n for n in nouns if len(n) >=2]

count=Counter(nouns)

wc_hong=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='white', mask=hong)
wc_hong=wc_hong.generate_from_frequencies(count)

plt.figure(figsize=(8,9))
plt.imshow(wc_hong, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from konlpy.tag import Hannanum
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = open("518연설문.txt", "r").read() #encoding='utf8'는 오류가 나는 경우 사용

engin=Hannanum()
nouns=engin.nouns(text)
nouns=[n for n in nouns if len(n) >=2]

count=Counter(nouns)
tags=count.most_common(40)

wordcloud=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='white', width=1200, height=800).generate_from_frequencies(dict(tags))

fig=plt.figure()
plt.axis('off')
plt.imshow(wordcloud)
```
