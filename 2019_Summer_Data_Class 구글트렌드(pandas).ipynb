```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

f1=open('불매검색량.csv')

data1=csv.reader(f1)
next(data1)

time=[]
google=[]
for row in data1:
    google.append(int(row[1]))
    time.append(row[0])
    
a=pd.DataFrame({'google':google})
a.set_index(pd.to_datetime(time), inplace=True)
    
plt.plot(a['google']['2018':])
```

```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/HANCooljazzB.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
plt.rc('font', family=font_name)

f1=open('모나미검색량.csv')

data1=csv.reader(f1)
next(data1)
google1=[]
time=[]
for row in data1:
    google1.append(int(row[1]))
    time.append(row[0])
    
f1=open('불매검색량.csv')

data2=csv.reader(f1)
next(data2)
google2=[]
for row in data2:
    google2.append(int(row[1]))
    
a=pd.DataFrame({'모나미':google1,'불매':google2})
a.set_index(pd.to_datetime(time), inplace=True)

plt.plot(a['모나미'], label='모나미')
plt.plot(a['불매'], label='불매')
plt.title('Pandas를 이용한 구글트렌드 분석')
plt.legend()
```
