```python
import matplotlib.pyplot as plt
import csv
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
import numpy as np

f= open('지하철역별이용현황.csv')

data=csv.reader(f)
next(data)

line=['1호선', '2호선', '3호선', '4호선', '5호선', '6호선','7호선','8호선', '9호선','우이신설선']
col=['darkblue','g','orange', 'blue','purple','y','g','r','grey','y']
result=[[],[],[],[],[],[],[],[],[],[]]
station=[[],[],[],[],[],[],[],[],[],[]]
m=100
for row in data:
    for i in range(0,10):
        if line[i] == row[1]:
            result[i].append(int(row[4])+int(row[5]))
            station[i].append(row[3])
            
s=[]
s1=[]
for i in range(0,10):
    for j in range(0,len(result[i])):
        if result[i][j]>=np.max(result[i]):
            s.append(station[i][j])
        if result[i][j]<=np.min(result[i]):
            s1.append(station[i][j])
        
            
        
plt.figure(figsize=(8,7), dpi=300)
plt.boxplot(result, labels=line)
for i in range (0,10):
    plt.text(i+1, max(result[i])+100000,s[i], rotation=1, size=5, color=col[i])
for i in range (0,10):
    plt.text(i+0.8, min(result[i])-100000,s1[i], rotation=1, size=5, color=col[i])
plt.title('지하철 노선별 이용현황')
plt.show()
    
```
