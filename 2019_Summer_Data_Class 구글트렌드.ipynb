```python
import matplotlib.pyplot as plt
import csv

f1=open('모나미검색량.csv')

data1=csv.reader(f1)
next(data1)
google=[]
for row in data1:
    google.append(int(row[1]))
    
plt.plot(google)
plt.show()
    
    
```

```python
import matplotlib.pyplot as plt
import csv
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(10,4))

f1=open('모나미검색량.csv')

data1=csv.reader(f1)
next(data1)
google1=[]
for row in data1:
    google1.append(int(row[1]))
    
plt.plot(google1, label='모나미')


f1=open('불매검색량.csv')

data2=csv.reader(f1)
next(data2)
google2=[]
for row in data2:
    google2.append(int(row[1]))
    
plt.plot(google2, label='불매')

plt.title('구글트렌드 검색량 비교')
plt.legend()
plt.show()
```
